## æ•°æ®å¯è§†åŒ–&æ¢ç´¢æ€§åˆ†æâ€”â€”è¿›é˜¶

ç»˜åˆ¶kçº¿å›¾æˆ‘ä»¬é‡‡ç”¨mplfinanceä¸­çš„candlestick_ochl()æ–¹æ³•ï¼Œä»¥æ—¥kçº¿ä¸ºä¾‹ï¼Œæˆ‘ä»¬è¦å‡†å¤‡çš„æ•°æ®æœ‰(time, open, close, high, low)ã€‚å…¶ä¸­ï¼Œtimeä¸ºæ¯æ—¥çš„æ—¥æœŸï¼Œopenï¼Œcloseï¼Œhighï¼Œlowåˆ†åˆ«ä¸ºæ¯æ—¥å¼€ç›˜ã€æ”¶ç›˜ã€æœ€é«˜ã€æœ€ä½ä»·ã€‚

å€¼å¾—ä¸€æçš„æ˜¯ï¼Œè‚¡å¸‚çš„å¼€ã€æ”¶ç›˜æ—¶é—´åˆ†åˆ«ä¸ºä¸Šåˆ9:30ä»¥åŠä¸‹åˆ15:00ï¼Œè€Œæ ¹æ®ä¹‹å‰æŸ¥çœ‹çš„ç»“æœï¼Œå­˜åœ¨ç€åœ¨è¯¥æ—¶é—´èŒƒå›´å¤–çš„æ•°æ®ï¼Œåº”é¦–å…ˆå‰”é™¤ä¹‹ã€‚

ä¸ºä¿è¯ç¨‹åºçš„å®Œæ•´æ€§ï¼Œé¿å…å¤§å®¶å¿˜è®°ï¼Œè¿™é‡Œæˆ‘ä»¬ä»å¤´å¼€å§‹ã€‚

```python
# å¯¼å…¥å¿…è¦åº“
import pandas as pd  # æ•°æ®å¤„ç†
import datetime  # æ—¶é—´æ ¼å¼å¤„ç†
from matplotlib.pylab import date2num  # æ—¶é—´æ ¼å¼å¤„ç†
from matplotlib import pyplot as plt  # ç»˜å›¾
from mplfinance.original_flavor import candlestick_ochl  # ç»˜åˆ¶kçº¿å›¾
from matplotlib import ticker as mticker  # åˆ»åº¦å¤„ç†
from matplotlib import dates as mdates  # æ—¶é—´æ ¼å¼å¤„ç†

data = pd.read_csv('lq_test.csv')
data = data.dropna().reset_index().drop(columns='index')
raw_time = data.pop('Unnamed: 0')
```

é¦–å…ˆæˆ‘ä»¬è¦æŠŠå¼€ç›˜æ—¶é—´å¤–çš„æ—¶é—´å‰”é™¤æ‰ã€‚ä¸ºæ­¤é¦–å…ˆè¦æŠŠæ—¥æœŸå’Œæ—¶é—´åˆ†éš”å¼€æ¥ï¼Œå¹¶æŠŠæ—¶é—´çš„æ ¼å¼æ”¹æˆæ–¹ä¾¿æ¯”è¾ƒçš„å½¢å¼ã€‚

```python
date_times = []
dates = []
times = []
date_time_format = '%Y-%m-%d %H:%M:%S'  # åŸsträ¸­çš„æ—¥æœŸ-æ—¶é—´æ ¼å¼
date_format = '%Y-%m-%d'  # å¾…è½¬æ—¥æœŸæ ¼å¼
time_format = '%H:%M:%S'  # å¾…è½¬æ—¶é—´æ ¼å¼

# å°†strè½¬ä¸ºdatetime.datetime
for i in raw_time:
    date_times.append(datetime.datetime.strptime(i, date_time_format))

# å°†æ—¥æœŸä¸æ—¶é—´æ‹†å¼€ï¼Œä¸”æ­¤æ—¶çš„æ—¥æœŸä¸æ—¶é—´çš„ç±»å‹åˆå˜æˆstr
for date_time in date_times:
    dates.append(date_time.strftime(date_format))
    times.append(date_time.strftime(time_format))

# æŠŠåˆ†ç¦»å¹¶è°ƒæ•´æ ¼å¼çš„æ—¥æœŸå’Œæ—¶é—´å‚¨å­˜åœ¨dataä¸­
data['date'] = dates
data['time'] = times
data_ = data.copy()  # ä¸ºé¿å…æ±¡æŸ“æºæ•°æ®ï¼Œå°†æ•°æ®æ‹·è´è‡³æ–°çš„DataFrameä¸­è¿›è¡Œå¤„ç†ï¼Œcopy()æ–¹æ³•é»˜è®¤æ·±æ‹·è´ï¼Œä¹‹åæˆ‘ä»¬è¿˜ä¼šæåˆ°è¿™ä¸ªæ¦‚å¿µ
```

ä¸Šè¿°å°†strè½¬ä¸ºdatetime.datetimeæœ‰ä¸¤ä¸ªç›®çš„ï¼Œä¸€æ˜¯å°†æ—¥æœŸä¸æ—¶é—´åˆ†å¼€ï¼ŒäºŒæ˜¯è°ƒæ•´æ—¶é—´æ ¼å¼ï¼Œæ–¹ä¾¿åç»­æ¯”è¾ƒã€‚æˆ‘è‡ªå·±ç”¨pycharmè¯»å‡ºçš„æ—¥æœŸæ—¶é—´æ ¼å¼ä¸º

    date_time_format = '%Y/%m/%d %H:%M'  # å¦‚ï¼š2015/4/16 9:16

ä½†åœ¨æœ¬notebookä¸­ï¼Œæ—¶é—´æ ¼å¼å·²ç»ä¸ºæ–¹ä¾¿æ¯”è¾ƒçš„å½¢å¼ï¼Œå› æ­¤å¯ä»¥ç®€åŒ–æ“ä½œï¼Œç›´æ¥åˆ†å‰²æ–‡æœ¬ï¼š

    for i in raw_time:
        a, b = i.split(' ')
        dates.append(a)
        times.append(b)
        
æ¥ä¸‹æ¥æˆ‘ä»¬æŠŠéå¼€ç›˜æ—¶é—´çš„æ•°æ®å‰”é™¤æ‰ï¼š

```python
data_.drop(data_[(data_.time < '09:31:00') | (data_.time > '15:00:00')].index, inplace=True)  # æŠŠéå¼€ç›˜æ—¶é—´é€šè¿‡å­—ç¬¦ä¸²æ¯”è¾ƒå¤§å°å»é™¤
data_ = data_.reset_index().drop(columns='index')
data_.head()
```

æ­¤å¤„æ¯”è¾ƒçš„å¹¶éæ˜¯æ—¶é—´ï¼Œè€Œæ˜¯å­—ç¬¦ä¸²ã€‚

å­—ç¬¦ä¸²çš„å¤§å°è§„åˆ™ä¸ºï¼Œä»ç¬¬ä¸€ä¸ªå…ƒç´ å¼€å§‹æ¯”è¾ƒå¤§å°ï¼Œå¤§å°ç›¸åŒåˆ™ç»§ç»­æ¯”è¾ƒç¬¬äºŒä¸ªï¼Œä»¥æ­¤ç±»æ¨ã€‚å¦‚æœä¹‹å‰çš„æ—¶é—´æ ¼å¼ä¸ºâ€˜%H:%Mâ€™(å¦‚9:16)ï¼Œæ¯”è¾ƒå¤§å°çš„æ—¶å€™å°±ä¼šå› ä¸º'9' > '1'ï¼Œå¯¼è‡´9ç‚¹åçš„æ‰€æœ‰æ—¶é—´éƒ½æ¯”9ç‚¹å°ã€‚å› æ­¤è¦è½¬æ ¼å¼åå†è¿›è¡Œå­—ç¬¦ä¸²çš„æ¯”è¾ƒã€‚

ç°åœ¨æŠŠéœ€è¦çš„æ•°æ®æå–å‡ºæ¥ï¼š

```python
Open = data_['open'][0::240].reset_index().drop(columns='index')
Close = data_['close'][239::240].reset_index().drop(columns='index')
```

ä»¥openä¸ºä¾‹ï¼Œå–å‡ºdata_ä¸­çš„'open'åˆ—ï¼ŒæŒ‰\[start:end:step\]å¯¹æ•°æ®è¿›è¡Œåˆ‡ç‰‡ã€‚æ¯æ—¥çš„å¼€ç›˜ä»·ä¸ºæ¯æ—¥é¦–ä¸ªopenå€¼ï¼Œæ”¶ç›˜ä»·ä¸ºæ¯æ—¥æœ€åçš„closeå€¼ã€‚æ¯å¤©å¼€ç›˜æ—¶é—´ä¸º240åˆ†é’Ÿï¼Œè®¾æ­¥é•¿240ï¼Œopenå’Œcloseåˆ‡ç‰‡çš„èµ·ç‚¹åˆ†åˆ«ä¸º0ä¸239ï¼Œç›´è‡³æ•°æ®æœ€æœ«ã€‚

```python
High = data_[['high', 'date']].groupby('date').max().reset_index()
Low = data_[['low', 'date']].groupby('date').min().reset_index()
Dates = High['date']
```

highå’Œlowåˆ†åˆ«æ˜¯æ¯å¤©çš„æœ€é«˜æœ€ä½ä»·ï¼Œåˆ©ç”¨groupby()æ–¹æ³•ï¼ŒæŠŠæ•°æ®æŒ‰æ—¥æœŸåˆ†ç±»ï¼Œå¹¶å–å‡ºæœ€å¤§ã€æœ€å°å€¼ã€‚æ³¨æ„åˆ°åˆè¿›è¡Œäº†reset_index()ï¼Œæ˜¯æ–¹ä¾¿ç›´æ¥å–å‡ºæ—¥æœŸå¤‡ç”¨ã€‚

candlestick_ochl()å¯¹æ—¶é—´æ ¼å¼æœ‰è¦æ±‚ï¼Œè¿›è¡Œè½¬æ¢ï¼š

```python
plot_dates = []
for Date in Dates:
    plot_date = datetime.datetime.strptime(Date, date_format)  # å…ˆæŠŠæ—¥æœŸæ ¼å¼è½¬å›datetime.datetimeä»¥æ»¡è¶³date2num()çš„ç±»å‹è¦æ±‚
    plot_dates.append(date2num(plot_date))
```

æ¥ä¸‹æ¥æˆ‘ä»¬æ¥ç»˜åˆ¶ä¸Šä¸‹çš„å­å›¾ã€‚

ä¸Šå›¾ä¸ºRSI(Relative Strength Indexï¼Œç›¸å¯¹å¼ºå¼±æŒ‡æ ‡)ï¼Œæ˜¯æ ¹æ®ä¸€å®šæ—¶æœŸå†…ä¸Šæ¶¨ç‚¹æ•°å’Œä¸‹è·Œç‚¹æ•°ä¹‹å’Œçš„æ¯”ç‡åˆ¶ä½œå‡ºçš„ä¸€ç§æŠ€æœ¯æ›²çº¿ã€‚èƒ½å¤Ÿåæ˜ å‡ºå¸‚åœºåœ¨ä¸€å®šæ—¶æœŸå†…çš„æ™¯æ°”ç¨‹åº¦ã€‚è¯¥åˆ†ææŒ‡æ ‡çš„è®¾è®¡æ˜¯ä»¥ä¸‰æ¡çº¿æ¥åæ˜ ä»·æ ¼èµ°åŠ¿çš„å¼ºå¼±ï¼Œè¿™ç§å›¾å½¢å¯ä»¥ä¸ºæŠ•èµ„è€…æä¾›æ“ä½œä¾æ®ï¼Œéå¸¸é€‚åˆåšçŸ­çº¿å·®ä»·æ“ä½œã€‚

RSIçš„åŸç†ç®€å•æ¥è¯´æ˜¯ä»¥æ•°å­—è®¡ç®—çš„æ–¹æ³•æ±‚å‡ºä¹°å–åŒæ–¹çš„åŠ›é‡å¯¹æ¯”ï¼Œè­¬å¦‚æœ‰100ä¸ªäººé¢å¯¹ä¸€ä»¶å•†å“ï¼Œå¦‚æœ50ä¸ªäººä»¥ä¸Šè¦ä¹°ï¼Œç«ç›¸æŠ¬ä»·ï¼Œå•†å“ä»·æ ¼å¿…æ¶¨ã€‚ç›¸åï¼Œå¦‚æœ50ä¸ªäººä»¥ä¸Šäº‰ç€å–å‡ºï¼Œä»·æ ¼è‡ªç„¶ä¸‹è·Œã€‚


å¼ºå¼±æŒ‡æ ‡ç†è®ºè®¤ä¸ºï¼Œä»»ä½•å¸‚ä»·çš„å¤§æ¶¨æˆ–å¤§è·Œï¼Œå‡åœ¨0-100ä¹‹é—´å˜åŠ¨ï¼Œæ ¹æ®å¸¸æ€åˆ†é…ï¼Œè®¤ä¸ºRSIå€¼å¤šåœ¨30-70ä¹‹é—´å˜åŠ¨ï¼Œé€šå¸¸80ç”šè‡³90æ—¶è¢«è®¤ä¸ºå¸‚åœºå·²åˆ°è¾¾è¶…ä¹°çŠ¶æ€ï¼Œè‡³æ­¤å¸‚åœºä»·æ ¼è‡ªç„¶ä¼šå›è½è°ƒæ•´ã€‚å½“ä»·æ ¼ä½è·Œè‡³30ä»¥ä¸‹å³è¢«è®¤ä¸ºæ˜¯è¶…å–çŠ¶æ€ï¼Œå¸‚ä»·å°†å‡ºç°åå¼¹å›å‡ã€‚

RSIçš„è®¡ç®—å¦‚ä¸‹ï¼š

RSI=100Ã—RS/(1+RS) æˆ–è€…,RSI=100ï¼100Ã·(1+RS)

RS=Xå¤©çš„å¹³å‡ä¸Šæ¶¨ç‚¹æ•°/Xå¤©çš„å¹³å‡ä¸‹è·Œç‚¹æ•°

ä¸€èˆ¬çš„RSIå‘¨æœŸä¸º6æ—¥ï¼Œ12æ—¥ï¼Œ24æ—¥ã€‚

```python
def cal_rsi(df0, period=6):  # é»˜è®¤å‘¨æœŸä¸º6æ—¥
    df0['diff'] = df0['close'] - df0['close'].shift(1)  # ç”¨diffå‚¨å­˜ä¸¤å¤©æ”¶ç›˜ä»·çš„å·®
    df0['diff'].fillna(0, inplace=True)  # ç©ºå€¼å¡«å……ä¸º0
    df0['up'] = df0['diff']  # diffèµ‹å€¼ç»™up
    df0['down'] = df0['diff']  # diffèµ‹å€¼ç»™down
    df0['up'][df0['up'] < 0] = 0  # æŠŠupä¸­å°äº0çš„ç½®é›¶
    df0['down'][df0['down'] > 0] = 0  # æŠŠdownä¸­å¤§äº0çš„ç½®é›¶
    df0['avg_up'] = df0['up'].rolling(period).sum() / period  # è®¡ç®—periodå¤©å†…å¹³å‡ä¸Šæ¶¨ç‚¹æ•°
    df0['avg_down'] = abs(df0['down'].rolling(period).sum() / period)  # è®¡ç®—periodå¤©å†…è¯„ä»·ä¸‹è·Œç‚¹æ•°
    df0['avg_up'].fillna(0, inplace=True)  # ç©ºå€¼å¡«å……ä¸º0
    df0['avg_down'].fillna(0, inplace=True)  # ç©ºå€¼å¡«å……ä¸º0
    df0['rsi'] = 100 - 100 / (1 + (df0['avg_up'] / df0['avg_down']))  # è®¡ç®—RSI
    return df0  # è¿”å›åŸDataFrame
```

```python
plot_mat = cal_rsi(plot_mat)
```

```python
fig = plt.figure(facecolor='#07000d', figsize=(15, 10))
ax = plt.subplot2grid((6, 4), (1, 0), rowspan=4, colspan=4, facecolor='#07000d')
ax.plot(plot_mat.time[100:160].values, mov_avg_ten[100:160], '#e1edf9', label='10days', linewidth=1.5)  
ax.plot(plot_mat.time[100:160].values, mov_avg_thirty[100:160], '#4ee6fd', label='10days', linewidth=1.5)
candlestick_ochl(ax, plot_mat[100:160].values, width=0.6, colorup='#ff1717', colordown='#53c156')
ax.grid(True, color='w')
ax.xaxis.set_major_locator(mticker.MaxNLocator(10))
ax.yaxis.set_major_locator(mticker.MaxNLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.yaxis.label.set_color('w')
ax.spines['bottom'].set_color('#5998ff')
ax.spines['top'].set_color('#5998ff')
ax.spines['left'].set_color('#5998ff')
ax.spines['right'].set_color('#5998ff')
ax.tick_params(axis='y', colors='w')
ax.tick_params(axis='x', colors='w')
plt.ylabel('Stock Price and Volume', color='w')
ax_ = ax.twinx()
ax_.fill_between(plot_mat.time[100:160].values, 0, Volume.volume[100:160].values,
                facecolor='#00ffe8', alpha=0.4)
ax_.grid(False)
ax_.set_ylim(0, 4*Volume.volume[100:160].values.max())
ax_.spines['bottom'].set_color('#5998ff')
ax_.spines['top'].set_color('#5998ff')
ax_.spines['left'].set_color('#5998ff')
ax_.spines['right'].set_color('#5998ff')
ax_.tick_params(axis='y', colors='w')
ax_.tick_params(axis='x', colors='w')

# ç»˜åˆ¶RSIæ›²çº¿

ax0 = plt.subplot2grid((6, 4), (0, 0), sharex=ax, rowspan=1, colspan=4, facecolor='#07000d')  # ç¬¬1è¡Œç¬¬1åˆ—èµ·ç”»ï¼Œå 1è¡Œ1åˆ—
col_rsi = '#c1f9f7'  # RSIæ›²çº¿çš„é¢œè‰²
col_pos = '#8f2020'  # ä¸Šè¾…åŠ©çº¿åŠå…¶å¡«å……è‰²
col_neg = '#386d13'  # ä¸‹è¾…åŠ©çº¿åŠå…¶å¡«å……è‰²
ax0.plot(plot_mat.time[100:160].values, plot_mat.rsi[100:160].values, col_rsi, linewidth=1.5)  # RSIæ›²çº¿åŠå…¶é¢œè‰²ï¼Œçº¿å®½
ax0.axhline(70, color=col_pos)  # ä¸Šè¾…åŠ©çº¿åŠå…¶é¢œè‰²
ax0.axhline(30, color=col_neg)  # ä¸‹è¾…åŠ©çº¿åŠå…¶é¢œè‰²
ax0.fill_between(plot_mat.time[100:160].values, plot_mat.rsi[100:160].values, 70, where=(plot_mat.rsi.values[100:160] >= 70),
                 facecolors=col_pos)  # æŠŠRSIæ›²çº¿å¤§äºç­‰äº70çš„éƒ¨åˆ†å¡«å……ä¸ºçº¢è‰²
ax0.fill_between(plot_mat.time[100:160].values, plot_mat.rsi[100:160].values, 30, where=(plot_mat.rsi.values[100:160] <= 30),
                 facecolors=col_neg)  # æŠŠRSIæ›²çº¿å°äºç­‰äº30çš„éƒ¨åˆ†å¡«å……ä¸ºç»¿è‰²
ax0.set_yticks([30, 70])  # è®¾ç½®è¾…åŠ©çº¿çš„åˆ»åº¦
ax0.spines['bottom'].set_color("#5998ff")
ax0.spines['top'].set_color("#5998ff")
ax0.spines['left'].set_color("#5998ff")
ax0.spines['right'].set_color("#5998ff")
ax0.tick_params(axis='x', colors='w')
ax0.tick_params(axis='y', colors='w')
plt.ylabel('RSI', color='w')

plt.plot()
```

æœ€åæ˜¯æœ€ä¸‹é¢çš„å­å›¾ï¼šMACDçº¿

MACD(Moving Average Convergence / Divergenceï¼Œå¼‚åŒç§»åŠ¨å¹³å‡çº¿)æ˜¯ä»åŒæŒ‡æ•°ç§»åŠ¨å¹³å‡çº¿å‘å±•è€Œæ¥çš„ï¼Œç”±å¿«çš„æŒ‡æ•°ç§»åŠ¨å¹³å‡çº¿ï¼ˆEMA12ï¼‰å‡å»æ…¢çš„æŒ‡æ•°ç§»åŠ¨å¹³å‡çº¿ï¼ˆEMA26ï¼‰å¾—åˆ°å¿«çº¿DIFï¼Œå†ç”¨2Ã—ï¼ˆå¿«çº¿DIF-DIFçš„9æ—¥åŠ æƒç§»åŠ¨å‡çº¿DEAï¼‰å¾—åˆ°MACDæŸ±ã€‚MACDçš„æ„ä¹‰å’ŒåŒç§»åŠ¨å¹³å‡çº¿åŸºæœ¬ç›¸åŒï¼Œå³ç”±å¿«ã€æ…¢å‡çº¿çš„ç¦»æ•£ã€èšåˆè¡¨å¾å½“å‰çš„å¤šç©ºçŠ¶æ€å’Œè‚¡ä»·å¯èƒ½çš„å‘å±•å˜åŒ–è¶‹åŠ¿ï¼Œä½†é˜…è¯»èµ·æ¥æ›´æ–¹ä¾¿ã€‚MACDçš„å˜åŒ–ä»£è¡¨ç€å¸‚åœºè¶‹åŠ¿çš„å˜åŒ–ï¼Œä¸åŒKçº¿çº§åˆ«çš„MACDä»£è¡¨å½“å‰çº§åˆ«å‘¨æœŸä¸­çš„ä¹°å–è¶‹åŠ¿ã€‚

MACDåœ¨åº”ç”¨ä¸Šåº”å…ˆè¡Œè®¡ç®—å‡ºå¿«é€Ÿï¼ˆä¸€èˆ¬é€‰12æ—¥ï¼‰ç§»åŠ¨å¹³å‡å€¼ä¸æ…¢é€Ÿï¼ˆä¸€èˆ¬é€‰26æ—¥ï¼‰ç§»åŠ¨å¹³å‡å€¼ã€‚ä»¥è¿™ä¸¤ä¸ªæ•°å€¼ä½œä¸ºæµ‹é‡ä¸¤è€…ï¼ˆå¿«é€Ÿä¸æ…¢é€Ÿçº¿ï¼‰é—´çš„â€œå·®ç¦»å€¼â€ä¾æ®ã€‚æ‰€è°“â€œå·®ç¦»å€¼â€ï¼ˆDIFï¼‰ï¼Œå³12æ—¥EMAæ•°å€¼å‡å»26æ—¥EMAæ•°å€¼ã€‚å› æ­¤ï¼Œåœ¨æŒç»­çš„æ¶¨åŠ¿ä¸­ï¼Œ12æ—¥EMAåœ¨26æ—¥EMAä¹‹ä¸Šã€‚å…¶é—´çš„æ­£å·®ç¦»å€¼ï¼ˆ+DIFï¼‰ä¼šæ„ˆæ¥æ„ˆå¤§ã€‚åä¹‹åœ¨è·ŒåŠ¿ä¸­ï¼Œå·®ç¦»å€¼å¯èƒ½å˜è´Ÿï¼ˆ-DIFï¼‰ï¼Œæ­¤æ—¶æ˜¯ç»å¯¹å€¼æ„ˆæ¥æ„ˆå¤§ã€‚è‡³äºè¡Œæƒ…å¼€å§‹å›è½¬ï¼Œæ­£æˆ–è´Ÿå·®ç¦»å€¼è¦ç¼©å°åˆ°ä¸€å®šçš„ç¨‹åº¦ï¼Œæ‰çœŸæ­£æ˜¯è¡Œæƒ…åè½¬çš„ä¿¡å·ã€‚MACDçš„åè½¬ä¿¡å·ç•Œå®šä¸ºâ€œå·®ç¦»å€¼â€çš„9æ—¥ç§»åŠ¨å¹³å‡å€¼ï¼ˆ9æ—¥DIFï¼‰ã€‚ åœ¨MACDçš„å¼‚åŒç§»åŠ¨å¹³å‡çº¿è®¡ç®—å…¬å¼ä¸­ï¼Œéƒ½åˆ†åˆ«åŠ T+1äº¤æ˜“æ—¥çš„ä»½é‡æƒå€¼ï¼Œä»¥ç°åœ¨æµè¡Œçš„å‚æ•°12å’Œ26ä¸ºä¾‹ï¼Œå…¶å…¬å¼å¦‚ä¸‹ï¼š

é¦–å…ˆè®¡ç®—å‡ºå¿«é€Ÿç§»åŠ¨å¹³å‡çº¿ï¼ˆå³EMA1ï¼‰å’Œæ…¢é€Ÿç§»åŠ¨å¹³å‡çº¿ï¼ˆå³EMA2ï¼‰ï¼Œä»¥æ­¤ä¸¤ä¸ªæ•°å€¼ï¼Œæ¥ä½œä¸ºæµ‹é‡ä¸¤è€…ï¼ˆå¿«æ…¢é€Ÿçº¿ï¼‰é—´çš„ç¦»å·®å€¼ï¼ˆDIFï¼‰çš„ä¾æ®ï¼Œç„¶åå†æ±‚DIFçš„Nå‘¨æœŸçš„å¹³æ»‘ç§»åŠ¨å¹³å‡çº¿DEAï¼ˆä¹Ÿå«MACDã€DEMï¼‰çº¿ã€‚

ä»¥EMA1çš„å‚æ•°ä¸º12æ—¥EMA2çš„å‚æ•°ä¸º26æ—¥ï¼ŒDIFçš„å‚æ•°ä¸º9æ—¥ä¸ºä¾‹æ¥çœ‹çœ‹MACDçš„è®¡ç®—è¿‡ç¨‹

1ã€è®¡ç®—ç§»åŠ¨å¹³å‡å€¼ï¼ˆEMAï¼‰

12æ—¥EMAçš„ç®—å¼ä¸º

EMAï¼ˆ12ï¼‰=å‰ä¸€æ—¥EMAï¼ˆ12ï¼‰Ã—11/13+ä»Šæ—¥æ”¶ç›˜ä»·Ã—2/13

26æ—¥EMAçš„ç®—å¼ä¸º

EMAï¼ˆ26ï¼‰=å‰ä¸€æ—¥EMAï¼ˆ26ï¼‰Ã—25/27+ä»Šæ—¥æ”¶ç›˜ä»·Ã—2/27

2ã€è®¡ç®—ç¦»å·®å€¼ï¼ˆDIFï¼‰

DIF=ä»Šæ—¥EMAï¼ˆ12ï¼‰ï¼ä»Šæ—¥EMAï¼ˆ26ï¼‰

3ã€è®¡ç®—DIFçš„9æ—¥EMA

æ ¹æ®ç¦»å·®å€¼è®¡ç®—å…¶9æ—¥çš„EMAï¼Œå³ç¦»å·®å¹³å‡å€¼ï¼Œæ˜¯æ‰€æ±‚çš„MACDå€¼ã€‚ä¸ºäº†ä¸ä¸æŒ‡æ ‡åŸåç›¸æ··æ·†ï¼Œæ­¤å€¼åˆåDEAæˆ–DEMã€‚

ä»Šæ—¥DEAï¼ˆMACDï¼‰=å‰ä¸€æ—¥DEAÃ—8/10+ä»Šæ—¥DIFÃ—2/10ã€‚

è®¡ç®—å‡ºçš„DIFå’ŒDEAçš„æ•°å€¼å‡ä¸ºæ­£å€¼æˆ–è´Ÿå€¼ã€‚

ç”¨ï¼ˆDIF-DEAï¼‰Ã—2å³ä¸ºMACDæŸ±çŠ¶å›¾ã€‚

EMAçš„è®¡ç®—å¦‚ä¸‹ï¼š

![ema.svg](ema.svg)

```python
def cal_ema(df0, period, is_dea=False):  # DEAä¸EMAçš„è®¡ç®—æ–¹å¼ç›¸åŒï¼Œå°è£…åœ¨åŒä¸€ä¸ªå‡½æ•°ä¸­ï¼Œç”¨is_deaæ¥ç¡®è®¤æ˜¯å¦æ˜¯DEA
    for i in range(len(df0)):
        if not is_dea:
            if i == 0:
                df0.loc[i, 'ema'+str(period)] = df0.loc[i, 'close']  # EMAåˆå§‹å€¼ä¸ºå½“å¤©æ”¶ç›˜ä»·
            else:
                df0.loc[i, 'ema'+str(period)] = (2*df0.loc[i, 'close']+(period-1)*df0.loc[i-1, 'ema'+str(period)])/(period+1)  # æŒ‰å…¬å¼è®¡ç®—
            ema = df0['ema'+str(period)]
        else:
            if i == 0:
                df0.loc[i, 'dea'+str(period)] = df0.loc[i, 'dif']
            else:
                df0.loc[i, 'dea'+str(period)] = ((period-1)*df0.loc[i-1, 'dea'+str(period)]+2*df0.loc[i, 'dif']) / (period+1)
            ema = df0['dea'+str(period)]
    return ema
```

```python
def cal_macd(df0, short=12, long=26, m=9):
    short_ema = cal_ema(df0, short)  # è®¡ç®—12æ—¥EMA
    long_ema = cal_ema(df0, long)  # è®¡ç®—26æ—¥EMA
    df0['dif'] = short_ema - long_ema  # è®¡ç®—DIF
    dea = cal_ema(df0, m, is_dea=True)  # è®¡ç®—DEA
    df0['macd'] = 2 * (df0['dif'] - df0['dea'+str(m)])  # è®¡ç®—MACD
    return df0
```

åœ¨cal_ema()ä¸­ï¼Œä¹‹æ‰€ä»¥è¦åœ¨å‘½ååé¢åŠ ä¸€ä¸ª'+str(period)'ï¼Œæ˜¯å› ä¸ºpythonä¸­çš„èµ‹å€¼é»˜è®¤ä¸ºæµ…æ‹·è´ï¼Œåªå¤åˆ¶æŒ‡å‘æŸä¸ªå¯¹è±¡çš„æŒ‡é’ˆï¼Œè€Œä¸å¤åˆ¶å€¼ã€‚æˆ‘ä»¬åœ¨cal_macd()ä¸­ï¼Œåˆ†åˆ«è®¡ç®—äº†ä¸¤æ¬¡emaï¼Œå¦‚æœä¸åŠ '+str(period)'ï¼Œåˆ™ç¬¬äºŒæ¬¡è°ƒç”¨cal_ema()èµ‹å€¼ç»™long_ema()æ—¶ï¼Œç”±äºæ–°è®¡ç®—çš„emaè¦†ç›–äº†åŸæ¥çš„ï¼Œè€Œshort_emaçš„æŒ‡é’ˆæ²¡æœ‰å˜åŒ–ï¼Œå¯¼è‡´short_emaçš„å€¼ä¸long_emaç›¸åŒï¼Œåˆ™åé¢è®¡ç®—difå¯¹äºŒå€¼åšå·®å¾—åˆ°çš„ç»“æœå¿…ç„¶ä¸º0ã€‚

```python
plot_mat = cal_macd(plot_mat)
```
